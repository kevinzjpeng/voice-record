## Step 5: Transcribe to Cantonese

This cell will transcribe all uploaded audio files to Cantonese.

```python
def format_timestamp(seconds):
    """Format seconds to HH:MM:SS format."""
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    secs = int(seconds % 60)
    return f"{hours:02d}:{minutes:02d}:{secs:02d}"

def transcribe_audio(audio_path, model):
    """Transcribe audio file to Cantonese."""
    print(f"\n{'='*60}")
    print(f"Transcribing: {audio_path}")
    print(f"{'='*60}")
    
    # Transcribe with Cantonese/Chinese language specified
    result = model.transcribe(
        str(audio_path),
        language='zh',  # Chinese (includes Cantonese)
        task='transcribe',
        verbose=True
    )
    
    # Create transcript file
    transcript_path = Path(audio_path).stem + '_transcript.txt'
    
    with open(transcript_path, 'w', encoding='utf-8') as f:
        f.write(f"Transcript of: {audio_path}\n")
        f.write(f"Language: Cantonese/Chinese\n")
        f.write(f"{'='*60}\n\n")
        f.write(result['text'].strip())
        f.write("\n\n")
        f.write(f"{'='*60}\n")
        f.write("Detailed segments:\n\n")
        
        # Write detailed segments with timestamps
        for segment in result['segments']:
            start_time = format_timestamp(segment['start'])
            end_time = format_timestamp(segment['end'])
            text = segment['text'].strip()
            f.write(f"[{start_time} -> {end_time}] {text}\n")
    
    print(f"\nâœ“ Transcript saved to: {transcript_path}")
    print(f"\nðŸ“ Transcript preview:")
    print(f"{'-'*60}")
    print(result['text'][:500])
    if len(result['text']) > 500:
        print("...")
    print(f"{'-'*60}")
    
    return transcript_path

# Transcribe all uploaded audio files
transcript_files = []

for filename in uploaded.keys():
    try:
        transcript_path = transcribe_audio(filename, model)
        transcript_files.append(transcript_path)
    except Exception as e:
        print(f"\nâœ— Error transcribing {filename}: {str(e)}")

print(f"\n{'='*60}")
print(f"âœ“ Transcription complete!")
print(f"Successfully transcribed {len(transcript_files)}/{len(uploaded)} file(s)")
print(f"{'='*60}")
```

## Step 6: Download Transcripts

Download all transcript files to your computer.

```python
print("ðŸ“¥ Downloading transcript files...\n")

for transcript_file in transcript_files:
    if os.path.exists(transcript_file):
        print(f"Downloading: {transcript_file}")
        files.download(transcript_file)

print("\nâœ“ All transcripts downloaded!")
```

## Optional: View Transcript in Notebook

Run this cell to view the transcript content directly in the notebook.

```python
# Display all transcripts
for transcript_file in transcript_files:
    if os.path.exists(transcript_file):
        print(f"\n{'='*60}")
        print(f"File: {transcript_file}")
        print(f"{'='*60}\n")
        with open(transcript_file, 'r', encoding='utf-8') as f:
            print(f.read())
        print("\n")
```
