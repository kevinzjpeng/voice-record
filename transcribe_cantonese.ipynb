# ðŸŽ™ï¸ Voice Transcription - One-Click Setup

**Super simple - just 3 steps:**

1. **Select your language** â¬‡ï¸ (English, Cantonese, or Mandarin)
2. **Run the transcription cell** (takes 2-3 minutes)
3. **Upload your audio file** when prompted
4. **That's it!** Transcript auto-downloads when done

Supported formats: MP3, WAV, M4A, FLAC, OGG

ðŸ’¡ **Tip:** Go to Runtime â†’ Change runtime type â†’ GPU (optional, makes it faster)

````python
# Language Selection Dropdown
from ipywidgets import Dropdown, VBox, HTML
from IPython.display import display

print("Select transcription language:")

language_dropdown = Dropdown(
    options=[
        ('ðŸ‡¬ðŸ‡§ English', 'en'),
        ('ðŸ‡­ðŸ‡° Cantonese', 'yue'),
        ('ðŸ‡¨ðŸ‡³ Mandarin', 'zh')
    ],
    value='yue',
    description='Language:',
    style={'description_width': '100px'}
)

selected_language = language_dropdown.value
display(language_dropdown)

# Display selected language
def on_language_change(change):
    global selected_language
    selected_language = change['new']
    lang_name = {'en': 'English', 'yue': 'Cantonese', 'zh': 'Mandarin'}.get(selected_language, 'yue')
    print(f"âœ“ Language selected: {lang_name}")

language_dropdown.observe(on_language_change, names='value')
print(f"âœ“ Language selected: Cantonese")
````

````python
#!/usr/bin/env python3
"""
ONE-STEP CANTONESE TRANSCRIPTION
Setup + Upload + Transcribe + Download (automatic)
"""

print("\n" + "="*70)
print("ðŸŽ™ï¸  INSTALLING & SETTING UP WHISPER (ONE TIME)")
print("="*70 + "\n")

# Step 1: Install dependencies
print("ðŸ“¦ Installing dependencies...")
import subprocess
subprocess.run(['pip', 'install', '-q', 'openai-whisper'], check=True)
subprocess.run(['apt-get', '-qq', 'install', '-y', 'ffmpeg'], check=True)
print("âœ“ Dependencies installed\n")

# Step 2: Import libraries
print("ðŸ“š Importing libraries...")
import whisper
import os
from pathlib import Path
from google.colab import files
import torch
print("âœ“ Libraries ready\n")

# Step 3: Load model
print("ðŸ¤– Loading Whisper Large model...")
print("   (This takes ~1-2 minutes first time)\n")
model = whisper.load_model("large")
print("âœ“ Model loaded\n")

# Step 4: Upload audio
print("="*70)
print("ðŸ“¤ UPLOAD YOUR AUDIO FILE")
print("="*70)
print("Click 'Choose Files' button and select your audio file")
print("Supported: MP3, WAV, M4A, FLAC, OGG\n")

uploaded = files.upload()

if not uploaded:
    print("âš ï¸  No files uploaded. Please try again.")
else:
    print(f"\nâœ“ Uploaded: {list(uploaded.keys())}\n")
    
    # Step 5: Transcribe
    print("="*70)
    lang_name = {'en': 'English', 'yue': 'Cantonese', 'zh': 'Mandarin'}.get(selected_language, 'Unknown')
    print(f"ðŸŽ™ï¸  TRANSCRIBING TO {lang_name.upper()} (wait 2-5 minutes)")
    print("="*70 + "\n")
    
    def format_timestamp(seconds):
        hours = int(seconds // 3600)
        minutes = int((seconds % 3600) // 60)
        secs = int(seconds % 60)
        return f"{hours:02d}:{minutes:02d}:{secs:02d}"
    
    for filename in uploaded.keys():
        print(f"\nðŸ“ Processing: {filename}")
        print("-" * 70)
        
        # Transcribe using selected language
        result = model.transcribe(
            filename,
            language=selected_language,
            task='transcribe',
            verbose=False
        )
        
        # Create transcript file
        transcript_path = Path(filename).stem + '_transcript.txt'
        
        # Get language name for display
        lang_name = {'en': 'English', 'yue': 'Cantonese', 'zh': 'Mandarin'}.get(selected_language, 'Unknown')
        
        with open(transcript_path, 'w', encoding='utf-8') as f:
            f.write(f"Transcript: {filename}\n")
            f.write(f"Language: {lang_name}\n")
            f.write("=" * 70 + "\n\n")
            f.write("Detailed segments:\n\n")
            
            for segment in result['segments']:
                start = format_timestamp(segment['start'])
                end = format_timestamp(segment['end'])
                text = segment['text'].strip()
                f.write(f"[{start} â†’ {end}] {text}\n")
        
        print(f"âœ“ Transcript created: {transcript_path}\n")
        
        # Show preview
        print("ðŸ“‹ Preview:")
        print("-" * 70)
        preview = result['text'][:300]
        print(preview)
        if len(result['text']) > 300:
            print("...(more)")
        print("-" * 70)
        
        # Step 6: Auto-download
        print(f"\nðŸ“¥ Downloading transcript...")
        files.download(transcript_path)
        print(f"âœ“ Downloaded: {transcript_path}\n")

print("\n" + "="*70)
print("ðŸŽ‰ ALL DONE!")
print("="*70)
print("Your transcript is downloaded to ~/Downloads/\n")
print("Next step: Run this in terminal to organize:")
print("   python3 scripts/trigger_colab.py --organize")
print("="*70 + "\n")
````
